## Let's create new columns to add in our split address

```python
# creating a new empty column to insert our splits
uni.loc[:,'street address'] = ""
uni.loc[:,'city'] = ""
uni.loc[:,'state'] = ""
uni.loc[:,'zipcode'] = ""
```

## Now let's split our address and add them into our new columns

```python
uni['street address'] = uni['address'].str.split('\,').str[-3].str.strip()
uni['city'] = uni['address'].str.split('\,').str[-2].str.strip()
uni['state'] = uni['address'].str.split(' ').str[-2].str.strip()
uni['zipcode'] = uni['address'].str.split(' ').str[-1].str.strip()

```

```python
uni.head()
```

## Now that we've split the address column we will delete the old one. 

```python
del uni['address']
```

```python
uni.head()
```

## Before moving on, we will re-organize the columns

```python
column_names = ["name", "Website", "street address","city","state","zipcode", "Type", "Size"]
uni = uni.reindex(columns=column_names)
```

```python
uni.head()
```

## Lets clean up our Type column 

```python
# Here we'll seperate whether or not the college is a public college or private.
uni['private'] = np.where(uni['Type'].str.contains("Private", "Public"), 'Private', 'Public')
# Here we'll seperate the types of funding

Type = [uni['Type'] == "4-year, Private not-for-profit",
        uni['Type'] == "4-year, primarily associate's, Public", 
        uni['Type'] == "4-year, primarily associate's, Private for-profit",
        uni['Type'] == "4-year, Private for-profit",
        uni['Type'] == "4-year, Public"]
funding_type = ['not-for-profit', 'not-for-profit', 'for profit', 'for profit', 'not-for-profit']
uni['funding'] = np.select(Type, funding_type)
```

```python
# Check to see how it looks
uni.head(10)
```

## Finally let's seperate whether or not a University primarily offers Associates or not

```python
uni['Associates (y=1/n=0)'] = np.where(uni['Type'].str.contains("primarily associate's"), '1', '0')
```

```python
# Check to see how it looks
uni.head(10)
```

## Now that the table is clean, let's save this to a new CSV
